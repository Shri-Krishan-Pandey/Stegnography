```python
# Image Steganography using ASCII-XOR Encryption
# This program hides secret text in an image using ASCII-XOR operation with a key

# Import necessary libraries
import cv2 # For image processing
import numpy as np # For array operations
import os
import matplotlib.pyplot as plt # For displaying images
import string
```

```python
# Create dictionaries for ASCII conversion
# d: character to ASCII value mapping
# c: ASCII value to character mapping
d = {chr(i):i for i in range(255)}
c = {i:chr(i) for i in range(255)}
```

```python
# Load the cover image
image_path = r"D:\Sublime Text, Python\Stegnography\6.jpg"
x=cv2.imread(image_path) # Read image in BGR format
x_rgb = cv2.cvtColor(x,cv2.COLOR_BGR2RGB) # Convert to RGB for display

# Display the original image
plt.imshow(x_rgb)
plt.title("Original image")
plt.axis("off")
plt.show()

# Print image dimensions (Height, Width, Channels)
print("image shape (Height, Width, Channels):",x.shape)
```

```python
# Define encryption key and secret text
key = "ImageKey"
text = "secret"
print("Key : ", key)
print("Text : ", text)
```

```python
# Convert text and key to ASCII values
text_ascii=[d[ch] for ch in text]
key_ascii=[d[ch] for ch in key]

print("Text ASCII", text_ascii)
print("Key ASCII", key_ascii)
```

```python
# Create a copy of the image for encryption
x_enc = x.copy()

# Initialize variables for pixel traversal
n,m,z = 0, 0, 0 # n: row, m: column, z: channel (0=B, 1=G, 2=R)
kl = 0 # Key index

l=len(text) # Length of secret text

# EMBEDDING PROCESS (Encryption)
for i in range(l):

    # Store original pixel value
    orig_val = x_enc[n,m,z]

    # XOR operation between text character and key character
    new_val = d[text[i]] ^ d[key[kl]]
    x_enc[n,m,z] = new_val # Modify the pixel value

    # Print embedding details
    print(f"Embedding '{text[i]}' (ASCII {d[text[i]]}) XOR '{key[kl]}' (ASCII {d[key[kl]]} = {new_val} at pixel ({n},{m},{z}) [Original = {orig_val}]") 

    # Move to next pixel position (diagonal movement)
    n += 1 # Next row
    m += 1 # Next column
    z = (z+1)%3 # Cycle through channels (B->G->R->B...)
    m = (m+1)%3 # Additional column movement for security
    kl = (kl+1)%len(key) # Cycle through key characters
    
```

```python
# Save and display the encrypted image
cv2.imwrite("encrypted_img.jpg", x_enc)
plt.imshow(cv2.cvtColor(x_enc,cv2.COLOR_BGR2RGB))
plt.title("Encrypted Image")
plt.axis('off')
plt.show()
```

```python
# DECRYPTION PROCESS
n, m, z = 0, 0, 0 # Reset to starting position
kl = 0 # Reset key index
decrypt = "" # To store decrypted text
for i in range(l):

    # Get the encrypted pixel value
	val = x_enc[n,m,z]

    # XOR operation with key to get original character
	orig_char = c[val ^ d[key[kl]]]
	decrypt += orig_char # Build the decrypted string

    # Print decryption details
	print(f"Decrypting pixel ({n},{m},{z}): {val} XOR {d[key[kl]]} = {val ^ d[key[kl]]} -> '{orig_char}'")

    # Move to next pixel position (same as embedding path)
	n += 1
	m +=1
	m = (m+1)%3
	z = (z +1) % 3
	kl = (kl + 1) % len(key)

# Print final decrypted text
print("Decrypted text:", decrypt)
```
